```python
from bloqade.tsim import Circuit
from bloqade import squin
```

```python
@squin.kernel
def main():
    q = squin.qalloc(6)
    squin.reset(q[0])
    squin.h(q[0])
    squin.t(q[0])
    squin.h(q[0])
    for i in range(5):
        if i % 2 == 0:
            squin.cx(q[0], q[i + 1])
        else:
            squin.cx(q[i + 1], q[0])
    squin.broadcast.measure(q)
```

```python
circuit = Circuit(main)
```

```python
circuit.diagram(height=200)
```

```python
circuit.diagram("pyzx", scale=30);
```

```python
sampler = circuit.compile_sampler()
sampler.sample(shots=100_000)
```
